The **Five-Number Summary** is a quick way to describe a dataset using **five key statistics**.  
It helps understand the **center, spread, and overall distribution**.

## **The Five Numbers**

1. **Minimum** â€“ The smallest value in the dataset  
2. **Q1 (First Quartile)** â€“ The median of the lower half (25th percentile)

---

## **Why Use It**

- Summarizes the **distribution** quickly  
- Helps detect **outliers**  
- Forms the basis for a **boxplot**

---

## **Example**

Data: 3, 5, 7, 8, 12, 15, 18

| Statistic | Value |
|-----------|-------|
| Minimum   | 3     |
| Q1        | 5     |
| Median    | 8     |
| Q3        | 15    |
| Maximum   | 18    |

## **Visualization**

The five-number summary is often visualized using a **boxplot**, which provides a clear picture of the distribution.

**Example Data:**  
3, 5, 7, 8, 12, 15, 18

### **Boxplot Components**

- **Box** = Q1 to Q3 (5 to 15)  
- **Median** = 8 (line inside the box)  
- **Whiskers** = Minimum (3) and Maximum (18)  
- **Outliers** = Any data points beyond the whiskers, plotted as separate points  

### **Explanation**

- The **box** represents the middle 50% of the data (from Q1 to Q3)  
- The **line inside the box** shows the **median (Q2)**  
- **Whiskers** extend to the **minimum and maximum** values  
- **Outliers**, if any, are plotted as individual points beyond the whiskers
