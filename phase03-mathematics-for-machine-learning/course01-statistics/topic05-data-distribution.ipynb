Data distribution shows **how values are spread** across a dataset. Understanding the distribution helps in choosing the right **statistical methods** and **visualizations**.

---

## **1. Types of Data Distribution**

1. **Uniform Distribution**  
   - All values occur **approximately equally**  
   - Example: Rolling a fair die  

2. **Normal Distribution (Bell Curve)**  
   - Symmetrical, most values cluster around the **mean**  
   - Mean = Median = Mode  
   - Example: Heights of adults  

3. **Skewed Distribution**  
   - **Asymmetrical** distribution  
   - **Positive skew**: Tail on the right (larger values)  
   - **Negative skew**: Tail on the left (smaller values)  

4. **Bimodal / Multimodal Distribution**  
   - Two or more **peaks** in the data  
   - Example: Test scores with two groups of students  

---

## **2. Visualizing Data Distribution**

- **Histogram**: Shows frequency of data in intervals  
- **Boxplot**: Shows five-number summary, median, spread, and outliers  
- **Frequency Polygon**: Line connecting midpoints of histogram bars  
- **Dot Plot / Stem-and-Leaf**: Small datasets, shows individual values  

---

## **3. Key Points**

- Helps understand **central tendency, spread, and skewness**  
- Guides choice of **statistical analysis**  
- Detects **outliers and unusual patterns**
