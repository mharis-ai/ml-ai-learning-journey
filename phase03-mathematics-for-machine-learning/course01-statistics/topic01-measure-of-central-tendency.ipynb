## **1. Mean (Arithmetic Average)**

The **mean** is the sum of all observations divided by the number of observations.

### Formula:
> Mean = Sum of numbers Ã· Total numbers

### Example:
Data: 4, 6, 8  
Mean = (4 + 6 + 8) / 3 = **6**

### When to Use:
- When data is **symmetrical**
- When there are **no extreme outliers**

## **2. Median (Middle Value)**

The **median** is the middle value when the data is arranged in ascending order.

### Steps:
1. Arrange data  
2. Pick the middle value  
   - If **n is odd** â†’ single middle value  
   - If **n is even** â†’ average of two middle values  

### Example:
Data: 3, 5, 7, 9  
Median = (5 + 7) / 2 = **6**

### When to Use:
- When data has **outliers**  
- For **skewed** distributions

---

## **3. Mode (Most Frequent Value)**

The **mode** is the value that appears **most often**.

### Example:
Data: 2, 3, 3, 5, 7  
Mode = **3**

### Types of Mode:
- **Unimodal** â†’ one mode  
- **Bimodal** â†’ two modes  
- **Multimodal** â†’ more than two modes  
- **No mode** â†’ all values occur once

### When to Use:
- For **categorical data**  
- When identifying the **most common** value

## **Summary Table**
