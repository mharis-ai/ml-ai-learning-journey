A **Z-Score** (or standard score) measures **how many standard deviations a data point is from the mean**.  
It helps to **compare values from different datasets** or detect **outliers**.

---

## **1. Formula**

For a **population**:

> Z = (X - Î¼) / Ïƒ

For a **sample**:

> Z = (X - XÌ„) / s

Where:  
- X = the data point  
- Î¼ = population mean  
- XÌ„ = sample mean  
- Ïƒ = population standard deviation  
- s = sample standard deviation

**Tip:** Use **population formula** when data represents the whole population, and **sample formula** when working with a subset.

---

## **2. Interpretation**

- Z > 0: Value is **above the mean**  
- Z < 0: Value is **below the mean**  
- |Z| > 3: Typically considered an **outlier**  

---

## **3. Example**

**Data (Sample):** 10, 12, 15, 18, 20  

- Mean bar{X} = 15  
- Standard Deviation (s) â‰ˆ 3.74  

**Z-Score Calculation for X = 20:**

> Z = (X - XÌ„) / s  

Z = (20 - 15) / 3.74 â‰ˆ 1.34

**Interpretation:** 20 is **1.34 standard deviations above the mean**.

---

## **4. Code Implementation**

```python
import numpy as np

data = [10, 12, 15, 18, 20]
mean = np.mean(data)
std = np.std(data)

z_scores = [(x - mean)/std for x in data]
print("Z-Scores:", z_scores)
```
