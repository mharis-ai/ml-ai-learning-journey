# ðŸ“Š Density Curve

A **density curve** is a **smooth curve** that represents the **distribution of a continuous random variable**.  
It shows the **probability of different outcomes** in a dataset.

---

## **1. Key Properties**

1. The total area under the curve = **1** (represents 100% probability)  
2. The curve is **never negative**  
3. The area under the curve over an interval represents the **probability** of the variable falling within that interval  
4. Mean and median may be at the **center** for symmetric distributions  

---

## **2. Types of Density Curves**

1. **Normal Distribution Curve (Bell Curve)**  
   - Symmetric, unimodal  
   - Mean = Median = Mode  
   - Example: Heights of adults  

2. **Skewed Density Curve**  
   - **Positive skew**: Tail on the right  
   - **Negative skew**: Tail on the left  

3. **Uniform Density Curve**  
   - All outcomes are equally likely  
   - Flat horizontal line  

---

## **3. Visualizing Density Curve with Seaborn**


```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [3, 5, 7, 8, 12, 15, 18]

sns.kdeplot(data, color='purple')
plt.title("Density Curve")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()
```
