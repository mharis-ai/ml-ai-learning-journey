## **1. Z-Test**

A **Z-test** is used when:

- Sample size **n > 30** (large sample)  
- Population standard deviation **Ïƒ is known**  
- Data is normally distributed or sample size is large (CLT)

### **Formula**
> Z = (XÌ„ âˆ’ Î¼â‚€) / (Ïƒ / âˆšn)

### **When to Use**
- Test a single population mean  
- Compare two population means  
- Test population proportions  

### **Example**
A company claims that the mean weight of sugar packets is **500g**.  
Sample of **40 packets**: Mean = 492g, Population SD = 30g, Î± = 0.05

**Hypotheses**  
- Hâ‚€: Î¼ = 500  
- Hâ‚: Î¼ â‰  500  

> Z = (492 âˆ’ 500) / (30 / âˆš40) = -1.68  

Critical Z for Î± = 0.05 (two-tailed) = Â±1.96  
Since **-1.68 âˆ‰ rejection region**, we **fail to reject Hâ‚€**.  

**Conclusion:** No significant evidence that packet weight differs from 500g.

---

### **One-Sample Z-Test Using P-Value Approach**

````python
import numpy as np 
from scipy.stats import norm 

sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]
population_mean = 170
population_std = 3
sample_mean = np.mean(sample)
n = len(sample)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))
alpha = 0.05

if p_value < alpha:
    print("I reject the null hypothesis, as the sample mean is significantly different from the population mean.")
else:
    print("I fail to reject the null hypothesis, as the sample mean is not significantly different from the population mean.")
````

## **2. t-Test**

A **t-test** is used when:

- Sample size **n â‰¤ 30**  
- Population SD **Ïƒ unknown**  
- Data is approximately normal

### **Types of t-tests**
- One-sample t-test â€“ compare sample mean to population mean  
- Independent two-sample t-test â€“ compare means of two independent groups  
- Paired t-test â€“ before/after or matched pairs  

### **Formula (One-Sample)**
> t = (XÌ„ âˆ’ Î¼â‚€) / (s / âˆšn)

### **Formula (Paired)**
> t = (dÌ„) / (s_d / âˆšn)

### **Example**
A trainer wants to check if a fitness program improves performance.

| Person | Before | After |
|--------|--------|-------|
| 1      | 20     | 25    |
| 2      | 22     | 27    |
| 3      | 19     | 23    |
| 4      | 21     | 24    |
| 5      | 20     | 26    |

Differences = [5, 5, 4, 3, 6], Mean difference = 4.6, SD = 1.14, n = 5  

> t = 4.6 / (1.14 / âˆš5) â‰ˆ 9.02  

df = 4, Critical t (Î±=0.05 two-tailed) â‰ˆ 2.776  
Since **9.02 > 2.776**, we **reject Hâ‚€** â†’ Program significantly improved performance.

---

### **One-Sample T-Test Using P-Value Approach**

````python
import numpy as np 
from scipy import stats 

sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]
sample_mean = np.mean(sample)
sample_std = np.std(sample,ddof = 1)
n = len(sample)
mean_population = 170

t_stats = (sample_mean - mean_population) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - stats.t.cdf(abs(t_stats),df = n -1 ))
alpha = 0.05

if p_value < alpha:
    print("I reject the null hypothesis, as the sample mean is significantly different from the population mean.")
else:
    print("I fail to reject the null hypothesis, as the sample mean is not significantly different from the population mean.")
````

### **Two-Sample Independent T-Test Using P-Value Approach**

````python
import numpy as np 
from scipy import stats 

group_A = [85, 88, 90, 92, 87, 85, 89, 91, 86, 88] 
group_B = [82, 84, 80, 83, 81, 79, 78, 85, 84, 83]

t_stats , p_value = stats.ttest_ind(group_A, group_B , equal_var= False)
alpha = 0.05

if p_value < alpha:
    print("I reject the null hypothesis, as there is a significant difference between the two groups.")
else:
    print("I fail to reject the null hypothesis, as there is no significant difference between the two groups.")
````

## **4. Chi-Square (Ï‡Â²) Test**

Used for **categorical data**.

### **Formula**
> Ï‡Â² = Î£ ((O âˆ’ E)Â² / E)

Where:  
- **O** = Observed frequency  
- **E** = Expected frequency under null hypothesis  
- Î£ = Sum over all categories  

### **Purpose**
- Test **independence** between two categorical variables  
- Test **goodness of fit** for an expected distribution  

### **Steps to Solve**

1. **Collect Observed Values (O)**  
These are the counts you have from your data.

2. **Calculate Expected Values (E)**  
    > E = (Row Total Ã— Column Total) / Grand Total

3. **Compute (Oâˆ’E)Â² / E** for each cell  

4. **Sum all cells** â†’ Ï‡Â² statistic  

5. **Determine degrees of freedom (df)**

6. **Compare Ï‡Â² with critical value** from Ï‡Â² table using df and chosen Î± (usually 0.05)

---

### **Chi-Square Test of Independence Using P-Value Approach**

````python
import numpy as np 
import seaborn as sns 
from scipy.stats import chi2_contingency
import pandas as pd 

df = sns.load_dataset('titanic')

contingency_table = pd.crosstab(df['sex'],df['survived'])
chi2, p_value , dof , expected = chi2_contingency(contingency_table)
alpha = 0.05

if p_value < alpha:
    print("I reject the null hypothesis, as there is a significant relationship between gender and survival.")
else:
    print("I fail to reject the null hypothesis, as there is no significant relationship between gender and survival.")
````

## **5. ANOVA (Analysis of Variance)**

Used for **comparing means of three or more groups**.

### **Formula**
> F = MSB / MSW

Where:  
- **MSB** = Mean Square Between Groups = SSB / df_between  
- **MSW** = Mean Square Within Groups = SSW / df_within  
- **SSB** = Sum of Squares Between Groups  
- **SSW** = Sum of Squares Within Groups  
- **F** = Test statistic used to compare group means  

### **Purpose**
- Test whether **there is a significant difference** between group means  
- Used when comparing **three or more groups**  

### **Steps to Solve**

1. **State the hypotheses**  
    - **Hâ‚€:** All group means are equal  
    - **Hâ‚:** At least one group mean is different  

2. **Calculate group means and overall mean**  
    - Group mean = sum of observations in group / n  
    - Overall mean = sum of all observations / total number of observations  

3. **Calculate Sum of Squares Between (SSB)**  
    > SSB = Î£ n_i Ã— (group mean âˆ’ overall mean)Â²  

4. **Calculate Sum of Squares Within (SSW)**  
    > SSW = Î£ (each observation âˆ’ group mean)Â²  

5. **Determine degrees of freedom**  
    > df_between = k âˆ’ 1  

    > df_within = N âˆ’ k  

    Where k = number of groups, N = total observations  

6. **Calculate Mean Squares**  
    > MSB = SSB / df_between  
    
    > MSW = SSW / df_within  

7. **Calculate F-statistic**  
    > F = MSB / MSW  

8. **Compare F with critical value** from F-table using df_between, df_within and chosen Î± (usually 0.05)  

9. **Make a decision**  
    - If F > F_critical â†’ **Reject Hâ‚€**  
    - If F â‰¤ F_critical â†’ **Fail to reject Hâ‚€**

### **Example: ANOVA Test**

**Step 1: State the hypotheses**

### **One-Way ANOVA Using P-Value Approach**

````python
import seaborn as sns
import pandas as pd 
from scipy.stats import f_oneway

df = sns.load_dataset('titanic')

df = df[['age','pclass']].dropna()
df['pclass'].unique()

class_1 = df[df['pclass'] == 1]['age']
class_2 = df[df['pclass'] == 2]['age']
class_3 = df[df['pclass'] == 3]['age']

f_stats, p_value = f_oneway(class_1,class_2,class_3)
alpha = 0.05

if p_value < alpha:
    print("I reject the null hypothesis, as there is a significant difference in at least one passenger class.")
else:
    print("I fail to reject the null hypothesis, as there is no significant difference among the passenger classes.")
````

## **Key Points**

- **Z-test**: used when the population standard deviation (Ïƒ) is known or the sample size is large  
- **t-test**: used when Ïƒ is unknown and the sample size is small  
- **Two-sample t-test**: compares means of two independent groups  
- **Chi-square test**: tests relationships between categorical variables  
- **ANOVA**: compares means across **3 or more** groups  

## **Steps for All Hypothesis Tests**

1. State the null and alternative hypotheses  
2. Compute the test statistic  
3. Compute the test statistic and evaluate it using both the p-value and the critical value  
4. Make a decision (reject or fail to reject the null hypothesis)
