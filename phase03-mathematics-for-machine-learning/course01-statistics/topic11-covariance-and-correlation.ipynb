## **4. Key Points**

- Covariance shows **direction only**, magnitude depends on units.  
- Correlation standardizes covariance to **unitless [-1, 1]**, easier to interpret.  
- Both assume **linear relationship**.  
- Use correlation when you want **strength and direction**, covariance mainly for **mathematical computation** (e.g., portfolio variance).
