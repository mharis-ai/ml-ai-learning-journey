# üìä Hypothesis Testing Basics

Hypothesis testing is a fundamental concept in inferential statistics used to make decisions or draw conclusions about a population based on sample data. It allows us to test assumptions, compare groups, and determine whether observed patterns are statistically significant.

---

## **1. Descriptive vs Inferential Statistics**

### **Descriptive Statistics**
Summarize and describe the main features of a dataset.  
Examples: mean, median, mode, standard deviation, charts.

### **Inferential Statistics**
Make predictions or decisions about a population based on a sample.  
Used for: hypothesis testing, confidence intervals, regression, etc.

---

## **2. What is Hypothesis Testing?**

A systematic procedure to determine whether there is enough evidence in a sample to support a claim about a population.

### **Purpose**
- Evaluate assumptions  
- Determine statistical significance  
- Make data-driven decisions

---

## **3. Null and Alternative Hypotheses (H‚ÇÄ and H‚ÇÅ)**

In hypothesis testing, we start by assuming a position (H‚ÇÄ) and then checking if the sample provides enough evidence to reject it.

### **Null Hypothesis (H‚ÇÄ)**  
- The default assumption  
- States **no effect**, **no difference**, or **no relationship**  
- Example:  
  - *H‚ÇÄ: The mean weight of apples = 150g*

### **Alternative Hypothesis (H‚ÇÅ / Ha)**
- Competing claim we want to test  
- States **effect exists**, **difference exists**, or **relationship exists**  
- Example:  
  - *H‚ÇÅ: The mean weight of apples ‚â† 150g*

---

## **4. Significance Level (Œ±)**

The probability of rejecting the null hypothesis when it is actually true.

Typical values:
- 0.05 (most common)  
- 0.01  
- 0.10

### **Meaning**
> Œ± = 5% ‚áí 5% risk of rejecting a true null hypothesis.

---

## **5. Confidence Interval (CI)**

A range of values that is likely to contain the true population parameter.

Example:  
If 95% CI = (47.2, 53.8), we say:  
> ‚ÄúWe are 95% confident that the population mean lies between 47.2 and 53.8.‚Äù

### **Relationship with Œ±**
> Confidence Level = 1 - alpha
- Œ± = 0.05 ‚Üí 95% CI  
- Œ± = 0.01 ‚Üí 99% CI

---

## **6. Type I and Type II Errors**

### **Type I Error (False Positive)**
Rejecting H‚ÇÄ when it is actually true.  
Probability = Œ±  
Example: Saying a treatment works when it actually doesn‚Äôt.

### **Type II Error (False Negative)**
Failing to reject H‚ÇÄ when H‚ÇÄ is false.  
Probability = Œ≤  
Example: Saying a treatment doesn‚Äôt work when it actually does.

---

## **7. Four Outcomes of Hypothesis Testing (Confusion Matrix Style)**

|                         | **H‚ÇÄ True**       | **H‚ÇÄ False**        |
|-------------------------|------------------|-------------------|
| **Reject H‚ÇÄ**           | Type I Error (Œ±) | Correct Decision ‚úî |
| **Fail to Reject H‚ÇÄ**   | Correct Decision ‚úî | Type II Error (Œ≤) |

This table shows the possible results when performing any hypothesis test.

## **8. One-Tailed vs Two-Tailed Tests**

### **Two-Tailed Test**
Tests for any difference (‚â†).  
Used when deviation in **either direction** matters.  

- Hypothesis example:  
  - H‚ÇÄ: Œº = 50  
  - H‚ÇÅ: Œº ‚â† 50  

Graphically: two rejection regions (left and right tails).

### **One-Tailed Test**
Tests for a directional effect (> or <).  
Used when only one direction matters.

- **Right-tailed test:**  
  - H‚ÇÅ: Œº > 50  
- **Left-tailed test:**  
  - H‚ÇÅ: Œº < 50  

Graphically: one rejection region on either left or right tail.

---

## **9. Summary**

- **Descriptive statistics** summarize data; **inferential statistics** draw conclusions about populations.  
- Hypothesis testing compares two competing statements: **H‚ÇÄ vs H‚ÇÅ**.  
- **Significance level (Œ±)** controls Type I error.  
- **Confidence intervals** give a range of likely values.  
- **Type I (Œ±)** and **Type II (Œ≤)** errors represent incorrect decisions.  
- Hypothesis outcomes can be represented like a **confusion matrix**.  
- Tests can be **one-tailed** (directional) or **two-tailed** (non-directional).
