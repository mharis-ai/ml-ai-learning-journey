# üìä Bayes‚Äô Theorem in Probability

**Bayes‚Äô Theorem** is one of the most important concepts in probability and statistics.  
It allows us to **update prior beliefs** when new information (evidence) becomes available.

Bayes' Theorem connects **marginal**, **joint**, and **conditional** probabilities into one powerful formula.

---

## **1. What Is Bayes‚Äô Theorem?**

Bayes‚Äô Theorem describes the probability of an event **A** given that another event **B** has occurred:

> P(A|B) = (P(B|A) √ó P(A)) / P(B)

Where:

- **P(A)** = Prior probability (initial belief about A)  
- **P(B|A)** = Likelihood (probability of B happening if A is true)  
- **P(B)** = Total probability of evidence B  
- **P(A|B)** = Posterior probability (updated probability after seeing B)

---

## **2. Understanding Each Component**

### **üîπ Prior Probability ‚Äî P(A)**  
Your initial assumption before seeing evidence.  
Example: Probability a randomly selected email is spam.

---

### **üîπ Likelihood ‚Äî P(B|A)**  
Probability of observing the evidence B **if event A is true**.  
Example: Probability the email contains the word ‚ÄúFree‚Äù if it is spam.

---

### **üîπ Evidence ‚Äî P(B)**  
Overall probability of evidence B occurring (regardless of A).  
Example: Probability any email contains ‚ÄúFree‚Äù.

---

### **üîπ Posterior Probability ‚Äî P(A|B)**  
Updated probability after incorporating the evidence.  
Example: If an email contains ‚ÄúFree‚Äù, what is the updated probability that it is spam?

---

## **3. Expanded Formula (Law of Total Probability)**

Sometimes P(B) is not directly known, so we compute it as:

> P(B) = P(B|A)P(A) + P(B|A')P(A')

Where **A‚Ä≤** is the complement of A.

---

## **4. Example of Bayes‚Äô Theorem**

Suppose:
- 20% of emails are spam ‚Üí **P(A) = 0.20**
- 70% of spam emails contain ‚ÄúFree‚Äù ‚Üí **P(B|A) = 0.70**
- 10% of non-spam emails contain ‚ÄúFree‚Äù ‚Üí **P(B|A‚Ä≤) = 0.10**

### **Step 1: Compute evidence P(B)**  

> P(B) = (0.70)(0.20) + (0.10)(0.80) = 0.14 + 0.08 = 0.22

### **Step 2: Apply Bayes‚Äô Theorem**

> P(A|B) = (P(B|A) √ó P(A)) / P(B) = (0.70 √ó 0.20) / 0.22 = 0.636

### **üìå Final Result:**  
If an email contains the word *‚ÄúFree‚Äù*, there is a **63.6% probability** it is spam.

---

## **5. Why Bayes‚Äô Theorem is Important**

- Used in **machine learning**, **spam detection**, **medical diagnosis**, **NLP**, and **decision systems**
- Helps update beliefs using data  
- Forms the foundation for **Naive Bayes Classifier**

---

## **6. Summary**

| Concept | Meaning |
|--------|---------|
| **Prior (P(A))** | Belief before seeing evidence |
| **Likelihood (P(B\|A))** | Probability of evidence if A is true |
| **Evidence (P(B))** | Total probability of the observed evidence |
| **Posterior (P(A\|B))** | Updated probability after seeing evidence |

**Bayes‚Äô Theorem = Updating knowledge with new information.**
