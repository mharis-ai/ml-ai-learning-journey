## **1. Venn Diagrams in Probability**

A **Venn diagram** is a visual representation of events using circles inside a rectangle (the sample space). It helps illustrate intersections, unions, complements, and mutually exclusive events.

**Why Venn Diagrams Are Useful:**
- Visualize overlapping probabilities  
- Understand relationships between multiple events  
- Useful in solving probability and set theory problems
