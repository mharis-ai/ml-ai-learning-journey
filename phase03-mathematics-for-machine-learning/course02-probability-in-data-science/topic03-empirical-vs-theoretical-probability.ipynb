Probability measures the **likelihood of an event occurring**, ranging from 0 (impossible) to 1 (certain).

---

## **1. Classical (Theoretical) Probability**

- Based on **equally likely outcomes**
- Calculated **without experiments**, using logic or reasoning

**Formula:**

> P(E) = Number of favorable outcomes / Total number of possible outcomes


**Examples:**

Tossing a fair coin - Probability of heads:

>P(Heads) = 1/2 = 0.5

Rolling a fair six-sided die - Probability of getting 4:

> P(4) = 1/6 â‰ˆ 0.1667

**Key Points:**
- No experiments needed
- Assumes all outcomes are equally likely
- Often called theoretical probability

---

## **2. Empirical (Experimental) Probability**

- Based on actual experiments or observations
- Calculated using data collected from trials

**Formula:**

> P(E) = Number of times event occurs / Total number of trials

**Examples:**

Toss a coin 100 times, heads appeared 55 times:

> P(Heads) = 55/100 = 0.55

Roll a die 60 times, the number 4 appeared 10 times:

> P(4) = 10/60 â‰ˆ 0.167

**Key Points:**
- Requires experimental data
- Useful when outcomes may not be equally likely
- Gives approximate probability, improves with more trials

---

## **3. Comparison Table**

| Feature | Classical Probability | Empirical Probability |
|---------|----------------------|----------------------|
| Basis | Logic / Theory | Experiments / Observations |
| Data Required | Not required | Required |
| Accuracy | Exact (ideal conditions) | Approximate |
| Example | Coin toss, dice roll | Toss coin 100 times, record outcomes |

---

## **4. Summary**

- **Classical Probability** = Calculated from theory, assumes all outcomes equally likely
- **Empirical Probability** = Calculated from actual data/experiments, can vary from theoretical value
- **Both are used together**: theoretical probability predicts, and empirical probability verifies
