# ⭐ Important Terms in Probability

---

## **1. Random Experiment**
A process or action that leads to **one and only one outcome**, but the exact outcome **cannot be predicted** with certainty.

**Example:**  
- Rolling a die  
- Tossing a coin  
- Drawing a card

---

## **2. Trial**
A **single performance** of a random experiment.

**Example:**  
- Tossing a coin **one time**  
- Rolling a die **once**

---

## **3. Outcome**
The **result** of a trial or a random experiment.

**Example:**  
- Getting **Heads** when you toss a coin  
- Getting **4** when you roll a die

---

## **4. Sample Space (S)**
The set of **all possible outcomes** of a random experiment.

**Notation:**  
S = {all outcomes}

**Example:**  
- For a coin toss: **S = {H, T}**  
- For a die roll: **S = {1, 2, 3, 4, 5, 6}**

---

## **5. Event**
A **subset of the sample space**, meaning one or more outcomes we are interested in.

**Example:**  
- Getting an even number when rolling a die → **E = {2, 4, 6}**  
- Getting a red card from a deck → 26 possible outcomes
