## **The Method of Exhaustion**

The **Method of Exhaustion** is an ancient technique for finding areas and volumes by approximating a shape with simpler shapes whose errors become smaller and smaller. As the approximations improve, the remaining “gap” between the estimate and the true value is gradually **exhausted** — approaching the exact answer.

##### **Intuitive Idea**

The method works by **bounding a shape from above and below** with simpler shapes, such as polygons or rectangles, and refining these approximations so that the difference between them becomes arbitrarily small.  

This process allows mathematicians to **capture the true area or volume** of complex shapes without direct measurement, relying instead on increasingly precise estimates.

##### **Why It Works**

As the approximations are refined:

- The **error shrinks**  
- The **bounds tighten** around the true value  
- The shape becomes **sandwiched** between two nearly identical estimates  

This concept is the **precursor to limits** in modern calculus and forms the foundation of integration.

##### **Modern Analogy**

The Method of Exhaustion is conceptually identical to **Riemann sums**, where an area under a curve is approximated by dividing it into small shapes (rectangles), and the approximation improves as the divisions become finer.

##### **Summary**

> **The Method of Exhaustion approximates a shape from above and below using increasingly accurate simpler shapes until the remaining error becomes negligible — an early form of the limit concept in calculus.**

## **The Calculus of Infinitesimals**

In machine learning, **the calculus of infinitesimals** refers to the idea of analyzing how functions change by looking at **infinitely small (infinitesimal) variations** in their inputs. These tiny changes let us understand how a model’s loss responds to adjustments in its parameters — which is exactly what optimization algorithms rely on.

##### **In practical ML terms:**

- An **infinitesimal change** in a model parameter (like a weight) produces an **infinitesimal change** in the loss.  
- The ratio of these changes is the **derivative**, which tells us the direction and magnitude of steepest improvement.  
- Gradient-based methods (like **SGD**, **Adam**, and **backpropagation**) use these infinitesimal rates of change to update parameters efficiently.  
- Frameworks like **TensorFlow** and **PyTorch** simulate this through **automatic differentiation**, which computes these infinitesimal sensitivities exactly.

##### **In essence:**  
> Infinitesimal calculus provides the mathematical machinery that allows ML models to learn by measuring how tiny changes in inputs or parameters affect outputs, making optimization and training possible.

## **Limits**

In mathematics and machine learning, a **limit** describes the value a function approaches as its input gets arbitrarily close to a certain point. Understanding limits is essential for:

- **Continuity** of functions  
- **Derivatives**, which measure how functions change  
- **Gradients** in optimization, allowing algorithms like gradient descent to update model parameters effectively  

##### **Continuous Functions**

For a **continuous function**, the limit at a point matches the function value. There are no jumps, holes, or breaks.  

**Example:**  
$$
f(x) = x^2 \quad \text{is continuous everywhere, so} \quad \lim_{x \to a} f(x) = f(a)
$$

##### **Discontinuous Functions**

Limits behave differently for **discontinuous functions**, where the function may have jumps, holes, or infinite spikes.  

**Examples:**  

- **Jump Discontinuity:** The function suddenly changes value at a point  
$$
f(x) = \begin{cases} 
1, & x < 0 \\ 
2, & x \ge 0 
\end{cases}
$$  

- **Removable Discontinuity (hole):** The function is missing or misaligned at a point  
$$
f(x) = \frac{x^2 - 1}{x - 1}, \quad x \neq 1
$$  

- **Infinite Discontinuity:** The function grows without bound  
$$
f(x) = \frac{1}{x}, \quad x = 0
$$  

##### **In essence:**  
> Limits capture how functions **approach values** near a point. For continuous functions, the approach is smooth, while for discontinuous functions, limits may behave unpredictably due to jumps, holes, or spikes.

### **Visualizing Limits**

*Let's walk through some examples to see how limits work and how functions behave as they approach specific points.*

$$\lim_{x \to 1} \frac{x^2 - 1}{x - 1}$$

```python
def limit(x):
    y = (x**2 - 1)/(x - 1)
    return y
```

> The function cannot be evaluated at `x = 1`, as doing so would result in a **division by zero**.

```python
limit(2)
```

*As $x$ approaches 1, the value of the function $y$ approaches 2.*

```python
limit(0.9)
```

```python
limit(0.999)
```

```python
limit(1.1)
```

```python
limit(1.001)
```

```python
y = limit(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axvline(x=1, color='purple', linestyle='--')
plt.axhline(y=2, color='purple', linestyle='--')
ax.plot(x,y)
plt.show()
```

$$\lim_{x \to 0} \frac{\text{sin } x}{x}$$

```python
def sin_limit(x):
    y = np.sin(x)/x
    return y
```

> The function cannot be evaluated at `x = 0`, as doing so would result in a **division by zero**.

*As $x$ approaches 0, the value of the function $y$ approaches 1.*

```python
sin_limit(0.1)
```

```python
sin_limit(0.001)
```

```python
sin_limit(-0.1)
```

```python
sin_limit(-0.001)
```

```python
y = sin_limit(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-1, 2)
plt.axvline(x=0, color='purple', linestyle='--')
plt.axhline(y=1, color='purple', linestyle='--')
ax.plot(x,y)
plt.show()
```

$$ \lim_{x \to \infty} \frac{25}{x} $$

```python
def inf_limit(x):
    y = 25/x
    return y
```

> The function cannot be evaluated at `x = 0`, as doing so would result in a **division by zero**.

*As $x$ approaches $\infty$, the value of the function $y$ approaches 0.*

```python
inf_limit(1e3)
```

```python
inf_limit(1e6)
```

```python
y = inf_limit(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-300, 300)
ax.plot(x, y)
plt.show()
```

*Behavior of $f(x)$ = 25/x near x = 0 showing left-hand and right-hand limits*

```python
left_x = x[x<0]
right_x = x[x>0]
```

```python
left_y = inf_limit(left_x)
right_y = inf_limit(right_x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='gray', linestyle='--')
plt.axhline(y=0, color='gray', linestyle='--')
ax.plot(left_x, left_y, c='C0')
ax.plot(right_x, right_y, c='C0')
plt.ylim(-300, 300)
plt.show()
```
