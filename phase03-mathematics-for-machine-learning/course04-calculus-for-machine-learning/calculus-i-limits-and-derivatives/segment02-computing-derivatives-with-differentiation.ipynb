## **The Delta Method**

The **Delta Method** is a technique used to approximate the derivative (or rate of change) of a function of a variable, especially when the exact derivative is difficult to compute. It uses the derivative of the function at a point to estimate small changes in the function based on small changes in the variable.

```python
x = np.linspace(-10, 10, 10000)
y = x**2 + 2*x + 2
```

```python
def f(x):
    y = x**2 + 2*x + 2
    return y
```

```python
y = f(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
ax.plot(x,y)
plt.show()
```

*Let's identify the slope where, say, $x = 2$*

*First, let's determine what $y$ is:*

```python
f(2)
```

*Let's call this point $P$, which is located at (2, 10):*

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
ax.plot(x,y)
plt.show()
```

> The _delta method_ uses the difference between two points to calculate slope. To illustrate this, let's define another point, $Q$ where, say, $x = 5$.

```python
f(5)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(5, 37, c = 'orange', zorder=3)
ax.plot(x,y)
plt.show()
```

*To find the slope $m$ between points $P$ and $Q$*

$$m = \frac{\text{change in }y}{\text{change in }x} = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1} = \frac{37-10}{5-2} = \frac{27}{3} = 9$$

```python
m = (37-10)/(5-2)
m
```

*To plot the line that passes through $P$ and $Q$, we can rearrange the equation of a line $y = mx + b$ to solve for $b$:*
$$b = y - mx$$

```python
b = 37-m*5
b
```

```python
line_y = m*x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(5, 37, c='orange', zorder=3)
plt.ylim(-5, 150) 
ax.plot(x, line_y, c='orange') 
ax.plot(x,y)
plt.show()
```

> The closer $Q$ becomes to $P$, the closer the slope $m$ comes to being the true tangent of the point $P$. Let's demonstrate this with another point $Q$ at $x = 2.1$.

*Previously, our $\Delta x$ between $Q$ and $P$ was equal to 3. Now it is much smaller:* $$\Delta x = x_2 - x_1 = 2.1 - 2 = 0.1 $$

```python
f(2.1)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(2.1, 10.61, c = 'orange', zorder=3)
ax.plot(x,y)
plt.show()
```

```python
m = (10.61-10)/(2.1-2)
m
```

```python
b = 10.61-m*2.1
b
```

```python
line_y = m*x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(2.1, 10.61, c='orange', zorder=3)
plt.ylim(-5, 150)
plt.plot(x, line_y, c='orange', zorder=3)
ax.plot(x,y)
plt.show()
```

> The closer $Q$ becomes to $P$ (i.e., $\Delta x$ approaches 0), the clearer it becomes that the slope $m$ at point $P$ = (2, 10) is equal to 6.

*Let's make $\Delta x$ extremely small, 0.000001, to illustrate this:*

```python
delta_x = 0.000001
delta_x
```

```python
x1 = 2
y1 = 10
```

*Rearranging $\Delta x = x_2 - x_1$, we can calculate $x_2$ for our point $Q$, which is now extremely close to $P$:*
$$x_2 = x_1 + \Delta x$$

```python
x2 = x1 + delta_x
x2
```

*$y_2$ for our point $Q$ can be obtained with the usual function $f(x)$:*
$$y_2 = f(x_2)$$

```python
y2 = f(x2)
y2
```

*To find the slope $m$, we continue to use* $$m = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1}$$

```python
m = (y2 - y1)/(x2 - x1)
m
```

> Using the delta method, we've shown that at point $P$, the slope of the curve is 6. 

```python
x1 = -1
```

```python
y1 = f(x1)
y1
```

*Point $P$ is located at (-1, 1)*

```python
delta_x
```

```python
x2 = x1 + delta_x
x2
```

```python
y2 = f(x2)
y2
```

> Quick aside: Pertinent to defining differentiation as an equation, an alternative way to calculate $y_2$ is $f(x + \Delta x)$

```python
y2 = f(x1 + delta_x)
y2
```

*Point $Q$ is at (-0.999999, 1.000000000001), extremely close to $P$.*

```python
m = (y2-y1)/(x2-x1)
m
```

*Therefore, as $x_2$ becomes infinitely close to $x_1$, it becomes clear that the slope $m$ at $x_1 = -1$ is equal to zero. Let's plot it out:*

```python
b = y2-m*x2
b
```

```python
line_y = m*x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(x1, y1)
plt.scatter(x2, y2, c='orange', zorder=3)
plt.ylim(-5, 150)
plt.plot(x, line_y, c='orange', zorder=3)
ax.plot(x,y)
plt.show()
```

As $Q$ becomes infinitely close to $P$:
* $x_2$ - $x_1$ approaches 0
* In other words, $\Delta x$ approaches 0
* This can be denoted as $\Delta x \to 0$

## **The Differentiation Equation**

*Using the delta method, we've derived the definition of differentiation from first principles. The derivative of $y$ (denoted $dy$) with respect to $x$ (denoted $dx$) can be represented as:*
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{\Delta y}{\Delta x}$$

> Expanding $\Delta y$ out to $y_2 - y_1$: 
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{y_2 - y_1}{\Delta x}$$

*Finally, replacing $y_1$ with $f(x)$ and replacing $y_2$ with $f(x + \Delta x)$, we obtain a common representation of differentiation:*
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}$$

*Let's observe the differentiation equation in action:*

```python
def f(x):
    y = x**2 + 2*x + 2
    return y
```

```python
def differentiation(f, x, delta):
    return (f(x + delta) - f(x)) / delta
```

```python
deltaX = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]
```

*As you decrease the values of $\Delta x$, the estimated slope gets closer to the actual slope ($m$) of the function.*

```python
for delta in deltaX:
    print(differentiation(f, 2, delta))
```

```python
for delta in deltaX:
    print(differentiation(f, -1, delta))
```

## **Rules of Differentiation**

The **Rules of Differentiation** are a set of standard formulas and principles used to find the derivative of a function. They provide systematic methods to compute derivatives for various types of functions efficiently, without having to use the limit definition each time.


- ### **Power Rule**

The **Power Rule** is used to differentiate functions of the form **f(x) = x**, where **n** is any real number.

##### **Formula**
$$
\frac{d}{dx}[x^n] = n \cdot x^{n-1}
$$

##### **Example**
Differentiate **f(x) = x^5**:

$$
f'(x) = 5 \cdot x^4
$$

- ### **Constant Multiple Rule**

The **Constant Multiple Rule** is used to differentiate a constant multiplied by a function.  

##### **Formula**
If $c$ is a constant and $f(x)$ is a differentiable function:

$$
\frac{d}{dx}[c \cdot f(x)] = c \cdot f'(x)
$$

##### **Example**
Differentiate $f(x) = 7x^3$:

$$
f'(x) = 7 \cdot 3x^2 = 21x^2
$$

- ### **Sum Rule**

The **Sum Rule** is used to differentiate the sum of two or more functions.  

##### **Formula**
If $f(x)$ and $g(x)$ are differentiable functions:

$$
\frac{d}{dx}[f(x) + g(x)] = f'(x) + g'(x)
$$

##### **Example**
Differentiate $f(x) = x^3 + 5x^2$:

$$
f'(x) = 3x^2 + 10x
$$


- ### **Product Rule**

The **Product Rule** is used to differentiate the product of two functions.  

##### **Formula**
If $u(x)$ and $v(x)$ are differentiable functions:

$$
\frac{d}{dx}[u(x) \cdot v(x)] = u'(x) \cdot v(x) + u(x) \cdot v'(x)
$$

##### **Example**
Differentiate $f(x) = x^2 \cdot \sin(x)$:

- Let $u(x) = x^2$, $v(x) = \sin(x)$  
- $u'(x) = 2x$, $v'(x) = \cos(x)$  

$$
f'(x) = 2x \cdot \sin(x) + x^2 \cdot \cos(x)
$$

- ### **Quotient Rule**

The **Quotient Rule** is used to differentiate the division of two functions.  

##### **Formula**
If $u(x)$ and $v(x)$ are differentiable functions:

$$
\frac{d}{dx}\left[\frac{u(x)}{v(x)}\right] = \frac{u'(x) \cdot v(x) - u(x) \cdot v'(x)}{[v(x)]^2}
$$

##### **Example**
Differentiate $f(x) = \frac{x^2}{\sin(x)}$:

- Let $u(x) = x^2$, $v(x) = \sin(x)$  
- $u'(x) = 2x$, $v'(x) = \cos(x)$  

$$
f'(x) = \frac{2x \cdot \sin(x) - x^2 \cdot \cos(x)}{\sin^2(x)}
$$

- ### **Chain Rule**

The **Chain Rule** is used to differentiate a composite function.  

##### **Formula**
If $y = f(g(x))$ and both $f$ and $g$ are differentiable:

$$
\frac{dy}{dx} = f'(g(x)) \cdot g'(x)
$$

##### **Example**
Differentiate $f(x) = (3x^2 + 2)^5$:

- Let $f(u) = u^5$, $u = 3x^2 + 2$  
- $f'(u) = 5u^4$, $u' = 6x$  

$$
f'(x) = 5(3x^2 + 2)^4 \cdot 6x = 30x(3x^2 + 2)^4
$$
