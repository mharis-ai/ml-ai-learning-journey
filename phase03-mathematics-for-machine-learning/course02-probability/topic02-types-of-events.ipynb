## **1. Simple Event**
An event that has **only one possible outcome**.

**Example:**  
- Rolling a die and getting **4**.  
- Drawing the **Ace of Spades**.

---

## **2. Compound Event**
An event that includes **two or more possible outcomes**.

**Example:**  
- Getting an **even number** when rolling a die â†’ {2, 4, 6}.  
- Drawing a **red card** from a deck (26 outcomes).

---

## **3. Independent Events**
Events where the **occurrence of one does not affect** the other.

**Example:**  
- Tossing two coins.  
  Getting heads on one coin does not affect the result of the second.

---

## **4. Dependent Events**
Events where the **occurrence of one affects** the other.

**Example:**  
- Drawing two cards **without replacement**.  
  The first draw changes the probability of the second.

---

## **5. Mutually Exclusive Events**
Events that **cannot occur at the same time**.

**Example:**  
- Rolling a die: You cannot get **3** and **5** in the same roll.  
- Drawing a card: You cannot draw a card that is **both a heart and a spade**.

---

## **6. Exhaustive Events**
A set of events is exhaustive if **together they include all possible outcomes**.

**Example:**  
When rolling a die:  
- Events {1, 2, 3} and {4, 5, 6} are exhaustive because they cover all six outcomes.

---

## **7. Impossible Event**
An event that **cannot happen**.  
**Example:** Rolling a die and getting **7**.

---

## **8. Certain Event**
An event that **will always happen**.  
**Example:** When tossing a coin, getting **heads or tails**.
