## **What Is Machine Learning?**

Machine Learning is the study of algorithms that enable computers to learn from data and improve their performance on a task without being explicitly programmed.

In traditional programming, a developer defines a fixed set of rules that map inputs to outputs. This approach is effective when the rules are clear and easy to specify. However, for many real-world problems, defining such rules is difficult or impractical. Machine learning addresses this limitation by allowing systems to learn the rules directly from data.

A machine learning system is trained using examples. From these examples, it learns a model that can make predictions or decisions on new, unseen data. As additional data becomes available, the model can continue to improve its performance without requiring changes to the underlying program logic.

### **Applications of Machine Learning**

Machine learning is particularly useful for problems in which the relationship between inputs and outputs is complex or cannot be described accurately using fixed rules.

In the context of this course, common applications include:

- **Predicting continuous values**, such as estimating house prices based on features like size  
- **Classifying inputs into categories**, such as determining whether an email is spam or not  
- **Learning patterns from data**, where manually designing rules would be difficult or inefficient  

These examples illustrate how machine learning systems can learn patterns from data and apply that knowledge to make accurate predictions on new examples.

### **Key Idea**

The central idea of machine learning in this course is learning from data to build models that generalize well and make accurate predictions, particularly in supervised learning scenarios.

## **Types of Machine Learning**

Machine learning can be broadly categorized into different types depending on the kind of data available and the task the algorithm is designed to perform. The main types are **Supervised Learning** and **Unsupervised Learning**.

### **Supervised Learning**

Supervised learning algorithms are trained using **labeled data**, where each input is paired with a known output. The goal is to learn a mapping from inputs to outputs that can be used to make predictions on new, unseen data. **The two primary types of supervised learning are: Regression, which predicts continuous numerical values, and Classification, which predicts discrete categories.**

##### **Regression**  
Regression involves predicting a continuous numeric value based on input features.  
*Example:* Predicting house prices from size and location.  

##### **Classification**  
Classification involves predicting a discrete category or class for each input.  
*Example:* Classifying emails as spam or not spam, or determining whether a patient has a particular disease.

Supervised learning is a core focus of this course because it allows us to **quantitatively evaluate model performance** by comparing predictions to known outputs.

### **Unsupervised Learning**

Unsupervised learning algorithms are trained using **unlabeled data**, where no output labels are provided. The goal is to **discover hidden patterns, structures, or relationships** in the data. **The main types of unsupervised learning tasks include Clustering, Anomaly Detection, and Dimensionality Reduction.**

##### **Clustering**  
Clustering involves grouping similar examples together based on their features.  
*Example:* Segmenting customers into groups based on purchasing behavior.  

##### **Anomaly Detection**  
Anomaly Detection identifies examples that deviate significantly from the majority of the data.  
*Example:* Detecting fraudulent transactions or unusual system behavior.  

##### **Dimensionality Reduction**  
Dimensionality Reduction reduces the number of input features while preserving important patterns in the data.  
*Example:* Simplifying high-dimensional image data for visualization or faster computation.

Unsupervised learning is particularly useful when labeled data is unavailable, allowing algorithms to **uncover insights and patterns** without predefined categories.

## **Univariate Linear Regression**

Univariate Linear Regression is a type of **supervised learning** where we predict a **single continuous output** from **one input feature**. It is one of the simplest models in machine learning and helps understand core concepts like predictions, cost, and gradient descent.

### **The Model**

The model predicts output using a straight line:

$h(x) = w * x + b$

Where:  
- `x` = input feature (e.g., size of a house in 1000 sq.ft)  
- `h(x)` = predicted output (e.g., house price in $1000s)  
- `w` = weight (slope of the line)  
- `b` = bias (intercept)

The goal is to find values of `w` and `b` that make predictions as close as possible to the actual data.

### **Example: Predicting House Prices**

Suppose we have the following dataset:

| Size (1000 sq.ft) | Price ($1000s) |
|------------------|----------------|
| 1.0              | 150            |
| 2.0              | 200            |
| 3.0              | 250            |
| 4.0              | 300            |
| 5.0              | 350            |

**Step 1: Start with a guess**  
- Initial `w=0`, `b=0`  
- Predicted prices: all 0  

**Step 2: Measure error**  
- Compare predicted prices to actual prices  
- Error shows how far off the line is  

**Step 3: Update parameters**  
- Use **gradient descent** to adjust `w` and `b` gradually  
- Move in the direction that reduces the error  

**Step 4: Make predictions**  
- Suppose after training we get `w=50` and `b=100`  
- Then for a house of 6 (1000 sq.ft):  

$h(6) = 50*6 + 100 = 400$


### **Key Points**

- Predicts output from **one input feature**  
- Uses a **line (w*x + b)** to fit data  
- **Gradient descent** is used to minimize prediction error  
- Foundation for more complex models like multivariate regression

## **Cost Function (Mean Squared Error)**

In linear regression, the **cost function** measures how well our model is performing by calculating the difference between the predicted outputs and the actual outputs. The goal is to **minimize this cost** to make accurate predictions.

### **The Formula**

For **univariate linear regression**, the cost function $J(w, b)$ is defined as:

$$
J(w, b) = \frac{1}{2m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})^2
$$

Where:  
- $m$ = number of training examples  
- $h(x^{(i)}) = w \cdot x^{(i)} + b$ = predicted value for the $i$-th example  
- $y^{(i)}$ = actual value for the $i$-th example  
- $\sum$ = sum over all training examples  

The factor $\frac{1}{2}$ is included to simplify the derivative during **gradient descent**.

### **Intuition**

- The cost function calculates the **average squared difference** between predicted and actual values.  
- A **high cost** means predictions are far from actual values.  
- A **low cost** means the model fits the data well.  

Think of it as a “score” for how bad our model is: lower is better.

### **Example Using the House Prices Dataset**

Suppose after an initial guess $w=0$, $b=0$, we have predictions:

| Size (1000 sq.ft) | Price ($1000s$) | Predicted Price $h(x)$ |
|------------------|-----------------|-----------------------|
| 1.0              | 150             | 0                     |
| 2.0              | 200             | 0                     |
| 3.0              | 250             | 0                     |
| 4.0              | 300             | 0                     |
| 5.0              | 350             | 0                     |

**Step 1: Compute the errors**: $h(x^{(i)}) - y^{(i)}$  

- Errors: -150, -200, -250, -300, -350  

**Step 2: Square the errors**  

- Squared Errors: 22500, 40000, 62500, 90000, 122500  

**Step 3: Compute the mean and divide by 2**:

$$
J(w, b) = \frac{1}{2 \cdot 5} \sum_{i=1}^{5} (h(x^{(i)}) - y^{(i)})^2 = \frac{1}{10} \cdot 337500 = 33750
$$

- The cost is **33750**, which is high, meaning the initial guess of $w$ and $b$ is poor.  
- By updating $w$ and $b$ using **gradient descent**, this cost will decrease.

### **Key Points**

- Measures how far off the predictions are from actual values  
- Always non-negative (0 is perfect)  
- Guides **gradient descent** in finding optimal $w$ and $b$  
- Lower cost → better fit of the line to data

## **Gradient Descent**

Gradient Descent is an **optimization algorithm** used to find the values of $w$ and $b$ that **minimize the cost function** $J(w, b)$ in linear regression. It works by updating the parameters iteratively in the direction that reduces the cost.

### **The Update Rule**

For each iteration, the parameters are updated as follows:

$$
w := w - \alpha \frac{\partial J(w, b)}{\partial w}
$$

$$
b := b - \alpha \frac{\partial J(w, b)}{\partial b}
$$

Where:  
- $\alpha$ = learning rate (controls the step size)  
- $\frac{\partial J(w, b)}{\partial w}$ = gradient of the cost with respect to $w$  
- $\frac{\partial J(w, b)}{\partial b}$ = gradient of the cost with respect to $b$

The gradients indicate **how much and in which direction** to adjust the parameters to reduce the cost.

### **Intuition**

- Gradient tells us the **slope of the cost function** at the current $(w, b)$  
- If the gradient is positive → decrease the parameter  
- If the gradient is negative → increase the parameter  
- Repeat until the cost stops decreasing or reaches a minimum

Think of it as **rolling downhill** to reach the lowest point of the cost function.

### **Example Using House Prices**

Suppose we have the following dataset:

| Size (1000 sq.ft) | Price ($1000s$) |
|------------------|----------------|
| 1.0              | 150            |
| 2.0              | 200            |
| 3.0              | 250            |
| 4.0              | 300            |
| 5.0              | 350            |

1. **Initial guess:** $w=0$, $b=0$  
2. **Compute gradients:**  

$$
\frac{\partial J}{\partial w} = \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)}) x^{(i)}
$$

$$
\frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})
$$

3. **Update parameters using learning rate $\alpha$**:

$$
w := w - \alpha \cdot \frac{\partial J}{\partial w}, \quad b := b - \alpha \cdot \frac{\partial J}{\partial b}
$$

4. **Repeat** until the cost $J(w, b)$ is minimized.  

After several iterations, the model learns values like $w=50$ and $b=100$, which provide accurate predictions for new house sizes.

### **Key Points**

- Iterative method to **minimize cost function**  
- Learning rate $\alpha$ controls the **size of each step**  
- Stops when **convergence** is reached (cost no longer decreases significantly)  
- Core algorithm for training **linear regression and other models**

## **Code Implementation of Univariate Linear Regression**

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Step 1: Define the dataset
# -----------------------------
# House sizes (1000 sq.ft)
x = np.array([1, 2, 3, 4, 5])
# House prices ($1000s)
y = np.array([150, 200, 250, 300, 350])

# Number of training examples
m = len(x)

# -----------------------------
# Step 2: Initialize parameters
# -----------------------------
w = 0.0  # weight (slope)
b = 0.0  # bias (intercept)
alpha = 0.01  # learning rate
iterations = 3501  # number of gradient descent steps

# -----------------------------
# Step 3: Define the cost function
# -----------------------------
def compute_cost(x, y, w, b):
    m = len(x)
    cost = (1/(2*m)) * np.sum((w*x + b - y)**2)
    return cost

# -----------------------------
# Step 4: Define gradient computation
# -----------------------------
def compute_gradient(x, y, w, b):
    m = len(x)
    dj_dw = (1/m) * np.sum((w*x + b - y) * x)
    dj_db = (1/m) * np.sum(w*x + b - y)
    return dj_dw, dj_db

# -----------------------------
# Step 6: Plot the results before gradient descent
# -----------------------------
plt.figure(figsize=(8,6))
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, w*x + b, color='red', label='Fitted line')
plt.xlabel('Size (1000 sq.ft)')
plt.ylabel('Price ($1000s)')
plt.title('Univariate Linear Regression before Gradient Descent')
plt.legend()
plt.show()

# -----------------------------
# Step 5: Perform gradient descent
# -----------------------------
for i in range(iterations):
    dj_dw, dj_db = compute_gradient(x, y, w, b)
    w = w - alpha * dj_dw
    b = b - alpha * dj_db
    if i % 100 == 0:  # print cost every 100 iterations
        print(f"Iteration {i:6d}: Cost {compute_cost(x, y, w, b):.2f}, w = {w:.2f}, b = {b:.2f}")

# -----------------------------
# Step 6: Plot the results after gradient descent
# -----------------------------
plt.figure(figsize=(8,6))
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, w*x + b, color='red', label='Fitted line')
plt.xlabel('Size (1000 sq.ft)')
plt.ylabel('Price ($1000s)')
plt.title('Univariate Linear Regression after Gradient Descent')
plt.legend()
plt.show()

# -----------------------------
# Step 7: Make predictions
# -----------------------------
new_house_size = 6  # 1000 sq.ft
predicted_price = w * new_house_size + b
print(f"Predicted price for a house of size {new_house_size} (1000 sq.ft): ${predicted_price:.2f}k")
```
