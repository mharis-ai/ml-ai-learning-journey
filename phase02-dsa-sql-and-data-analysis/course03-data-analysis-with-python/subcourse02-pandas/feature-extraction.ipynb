```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv(r'anime.csv')
```

# Anime Data Insights

##### Make a new column for episode count

```python
def extract_episodes(txt):
    data = ''
    check = False
    for i in txt:
        if i == ')':
            break
        if check:
            data += i
        if i == '(':
            check = True
    return data

df['Episodes'] = df['Title'].apply(extract_episodes)
df.head()
```

##### Make a new column for time stamp

```python
def extract_timestamp(txt):
    data = ''
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1,i+20):
                data += txt[j]
            return data
df['Timestamp'] = df['Title'].apply(extract_timestamp)
df.head()
```

##### Which anime has the highest score?

```python
df[df['Score'] == df['Score'].max()]
```

##### Which anime has the highest episode count?

##### Which is the longest running anime?

```python
from datetime import datetime

def count_months(date_range):
    start, end = date_range.split(" - ")

    # Convert both to datetime (assumes format 'Mon YYYY')
    s = datetime.strptime(start, "%b %Y")
    e = datetime.strptime(end, "%b %Y")

    # Convert to month counts
    return (e.year - s.year) * 12 + (e.month - s.month) + 1

df['Months'] = df['Timestamp'].apply(count_months)
df[df['Months'] == df['Months'].max()]
```
