```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

## Matplotlib Basics

```python
x = np.arange(1,11)
x
y = x ** 2
y
```

```python
plt.plot(x,y)
plt.title("GRAPH")
plt.xlabel("Natural Numbers")
plt.ylabel("Square of Natural Numbers")
```

## Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

### Object oriented way

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches
dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(20,8),dpi = 100)
axis1 = fig.add_axes([0.1,0.1,.4,.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```

## DESIGNING

```python
plt.plot(x,y,color = "red")
```

```python
x = np.linspace(1, 20, 20)
y1 = 2*x + 5           
y2 = x**2 / 5          
y3 = np.sqrt(x) * 10  

style1 = dict(color="royalblue", linewidth=3, marker='o',
              markersize=7, markerfacecolor="white", markeredgecolor="royalblue")

style2 = dict(color="crimson", linewidth=3, linestyle='--', marker='s',
              markersize=7, markerfacecolor="white", markeredgecolor="crimson")

style3 = dict(color="darkgreen", linewidth=3, linestyle='-', marker='D',
              markersize=7, markerfacecolor="white", markeredgecolor="darkgreen")

fig, ax = plt.subplots(figsize=(10,5))

ax.plot(x, y1, label="Linear: 2x + 5", **style1)
ax.plot(x, y2, label="Quadratic: x / 5", **style2)
ax.plot(x, y3, label="x  10", **style3)

ax.set_title("Multi-Line Style", fontsize=16, weight='bold')
ax.set_xlabel("X Axis", fontsize=12)
ax.set_ylabel("Y Axis", fontsize=12)

ax.legend(frameon=False, fontsize=12)
ax.grid(linestyle='--', alpha=0.3)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
